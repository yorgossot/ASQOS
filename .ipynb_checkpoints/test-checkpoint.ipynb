```python
from system import *
from qutip.qip.operations import rz
import sage.all as sg
sim.C_val_range = [10,100,1000,10000]
sim.AVG_P_failure = np.zeros(np.shape(sim.C_val_range))
sim.AVG_Infidelity = np.zeros(np.shape(sim.C_val_range))
sim.AVG_gate_time = np.zeros(np.shape(sim.C_val_range))
print(f'Starting qutip simulations for C in {sim.C_val_range[0]}-{sim.C_val_range[-1]}') 
for (C_val_i,C_val) in enumerate(sim.C_val_range):      

    g_val = sg.sqrt(C_val*sim.parameters['kappa_c'] *sim.parameters['gamma'] )

    eff_hamiltonian_num = symround( sim.eff_hamiltonian_C.subs(g=g_val).expand() ,digits=14,show_del=False)
    eff_hamiltonian_num = eff_hamiltonian_num.numpy().astype(float)
    H_obj = qt.Qobj(eff_hamiltonian_num)

    sim.H_obj = H_obj

    ExponentialFactorsOfT = [-1j*H_obj[i,i] for i in range(sim.setup.gs_dim)]
    LossFactorsOfT = [0 for i in range(sim.setup.gs_dim)]
    L_num = symround(sim.eff_lind_master_eq_C.subs(g=g_val).expand() ,digits=14,show_del=False)
    for lind_op in range(sim.lind_op_number):
        for which in range(sim.setup.gs_dim):
            Ln = L_num
            Loss = abs(complex(L_num))**2 
            ExponentialFactorsOfT[which] -= Loss 
            LossFactorsOfT[which] += Loss


    init_state = np.array([1,1,1,1])/2

    psif_gs  =  qt.Qobj(np.array([1,1,1,-1])/2)  #target state

    H = [eff_hamiltonian_num[diag,diag] for diag in range(sim.setup.gs_dim)]

    gate_time =  np.abs( np.pi  /(H[3]+H[0]-H[1]-H[2]) )

    PSuccess = 0
    for i in range(sim.setup.gs_dim):
        PSuccess += init_state[i]**2 * np.exp( - LossFactorsOfT[i]*gate_time )


    PureEvolutionVector = []   
    for i in range(sim.setup.gs_dim):
        Evolution = np.exp(ExponentialFactorsOfT[i]*gate_time)  
        if np.real(ExponentialFactorsOfT[i])>0:
            print(f"fail{np.real(Evolution)}")
        PureEvolutionVector.append( Evolution*init_state[i] )

    PureEvolutionVector = qt.Qobj( np.array(PureEvolutionVector) )

    PureEvolutionDM = qt.ket2dm( PureEvolutionVector ) 

    #post process rotations

    r1 = - gate_time * (H[0] - H[2])           
    R1 = qt.Qobj(qt.tensor( rz(r1 ),qt.identity(2)).full())


    r2 = -gate_time * (H[0] - H[1])
    R2 = qt.Qobj( qt.tensor(qt.identity(2),rz(r2 )).full())

    #simulate


    RotatedFinalState = R1.dag()*R2.dag()*psif_gs

    StateFidelity = qt.fidelity(PureEvolutionDM ,RotatedFinalState) /PSuccess
            

    f_prob = 1- PSuccess 
    infidelity = 1 - StateFidelity

    sim.AVG_Infidelity[C_val_i] = infidelity
    sim.AVG_P_failure[C_val_i] = f_prob
    sim.AVG_gate_time[C_val_i] = gate_time
```
